**Sales is highly corelated with the TV**

Let's train our model using linear regression as it is coorelated with only one variable TV


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['TV']], df[['Sales']], train_size=0.7, test_size=0.3, random_state=0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
res= model.predict(X_test)
print(res)
```

```python
model.coef_
```

```python
model.intercept_
```
